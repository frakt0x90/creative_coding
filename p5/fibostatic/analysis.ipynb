```python
import pandas
import numpy
from plotnine import ggplot, aes, geom_point, geom_density
from sympy.ntheory.factor_ import udivisor_count, primefactors, proper_divisor_count
```

```python
data = pandas.read_json('data.json').set_axis(['mod', 'period'], axis=1)
```

```python
(ggplot(data) + geom_point(aes('mod', 'period')))
```

```python
data = data[data.period < 50000]
```

```python
prime_facs = lambda x: len(primefactors(x))
data['udivisors'] = data['mod'].apply(udivisor_count)
data['divisors'] = data['mod'].apply(proper_divisor_count)
data['prime_factors'] = data['mod'].apply(prime_facs)
data['logperiod'] = numpy.log(data.period)
```

```python
(ggplot(data) + geom_point(aes('mod', 'logperiod')))
```

```python
(ggplot(data) + geom_point(aes('divisors', 'logperiod')))
```

```python
(ggplot(data) + geom_density(aes('logperiod', fill='factor(prime_factors)'), alpha=.2))
```

```python
(ggplot(data) + geom_density(aes('logperiod', fill='factor(divisors)'), alpha=.2))
```
